```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

#Analysis of literacy Rate and total crime 
1. Literacy Rate have been taken from https://www.careindia.org/blog/literacy-rate-in-india/
2. Total Crime have been added using sum if formula in excel

```python
df=pd.read_csv('qus 1.csv')
df.head()
```

```python
#checking Info
df.info()
```

```python
x= pd.DataFrame(df.groupby(['Literacy Rate'])['Total Crimes'].sum().reset_index())
x
```

```python
plt.bar(x['Literacy Rate'], df['Total Crimes'], color ='maroon',
        width = 0.4)
plt.xlabel("Literacy Rate")
plt.ylabel("Total Crimes")
plt.title("Literacy and Total crimes")
plt.show()
```

```python
#By plotting the above graph it has been been observed that higher the literacy rate lower will be crime 
#In some placed where literacy rate is higher but crime also lower due to small area and small U/T
```

```python

plt.figure(figsize=(20,10))
sns.set_style('whitegrid')
sns.pointplot(x='Literacy Rate', y='Total Crimes', data=df,join=False)
plt.title('Literacy and Total crimesn', fontsize=22, fontweight='bold')
plt.xlabel('Total Crimes',{'fontsize':15,'fontweight' :'bold'})
plt.ylabel('Total Crimes',{'fontsize':15,'fontweight' :'bold'})
plt.xticks(fontsize=16,fontweight ='bold')
plt.yticks(fontsize=16,fontweight ='bold')
plt.legend(fontsize=14)
```
