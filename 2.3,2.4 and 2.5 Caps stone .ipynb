# 2.4 Analysis of area vs overall crime

```python
#below data shows total crime V\S Area of states
df=pd.read_csv('area of states with crime.csv')
df
```

```python
#State and there area in accodance with percentage
df.groupby("STATE/UT")["Area of states"].sum().plot.pie(autopct='%1.0f%%')
plt.tight_layout()
```

```python
#Bar graph for each states and there area
plt.figure(figsize=(20,10))
df.groupby("STATE/UT")["Area of states"].sum().plot.bar()
```

```python
#comparision of row with columns 
sns.pairplot(df,palette='Dark2')
plt.show()
```

```python
#in above graphs it have been obseved that higer all three are co relation with each other higher the Arear higher the population and same go with crimes
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),linewidths=.2,fmt='.1g',linecolor='black',annot=True,cmap='Blues_r')
plt.yticks(rotation=0);
plt.show()
```

```python
#state/UT and Area of states --RJ is the highest area
plt.figure(figsize=(20,10))
sns.barplot(x=df['Area of states'], y=df['STATE/UT'])

```

```python
plt.figure(figsize=(20,10))
sns.barplot(x=df['Total Crimes'], y=df['STATE/UT'])

```
